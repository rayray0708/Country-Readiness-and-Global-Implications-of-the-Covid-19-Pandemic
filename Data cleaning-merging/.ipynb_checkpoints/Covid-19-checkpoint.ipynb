```python
countries = json['stats']['breakdowns']
countries
```

```python
#Convert the JSON object into a flattened Pandas dataframe using .json_normalize()
countries_flattened = pd.json_normalize(countries, max_level = 1)
countries_flattened
```

```python
#Check the type of countries_flattened
type(countries_flattened)
```

```python
#Rename all columns
countries_flattened = countries_flattened.rename(columns={'totalConfirmedCases': 'Total Confirmed Cases', 
                                                  'newlyConfirmedCases': 'Newly Confirmed Cases', 
                                                  'totalDeaths': 'Total Deaths',
                                                    'newDeaths': 'New Deaths',
                                                    'totalRecoveredCases':'Total Recovered Cases',
                                                    'newlyRecoveredCases': 'Newly Recovered Cases',
                                                    'location.long':'Lon',
                                                    'location.countryOrRegion':'Country/Region',
                                                    'location.provinceOrState':'Province/State',
                                                    'location.county':'County',
                                                    'location.isoCode':'ISO Code',
                                                    'location.lat': 'Lat'})
countries_flattened.head()
```

```python
#Reorder columns
countries_flattened = countries_flattened[['Country/Region',
                                          'Total Confirmed Cases',
                                          'Newly Confirmed Cases',
                                          'Total Deaths',
                                          'New Deaths',
                                          'Total Recovered Cases',
                                          'Newly Recovered Cases',
                                          'Province/State',
                                          'County',
                                          'ISO Code',
                                          'Lat',
                                          'Lon']]
countries_flattened.head()
```

```python
#Check if columns 'Province/State' and 'County' contain any other values apart from 'None'
countries_flattened['Province/State'].unique()
countries_flattened['County'].unique()
```

```python
# Drop 'Province/State' and 'County' columns
countries_flattened = countries_flattened.drop(['Province/State','County'], axis = 1)
countries_flattened.head()
```

```python
from pathlib import Path
import pandas as pd
```

```python
#Set path to the GDP CSV file
csv_reader = Path("../Resources/filtered_GDP_health_expenditure_final.csv")
csv_reader
```

```python
# Set path to the vaccination CSV file
csv_reader1 = Path("../Resources/vaccination-data.csv")
csv_reader1
```

```python
# Read the GDP CSV 
GDP_data = pd.read_csv(csv_reader,low_memory=False)
GDP_data.head(100)
```

```python
# Read the vaccination CSV
vaccination_df = pd.read_csv(csv_reader1,low_memory=False)
vaccination_df.head(100)
```

```python
#Select the desired columns from the GDP_data
GDP_df = GDP_data[["Country Name", "2015","2016","2017", "2018", "2019"]]
GDP_df
```

```python
GDP_df.info()
vaccination_df.info()
countries_flattened.info()
```

```python
# Merge the vaccination with countries dataframes
vaccination_countries_df = vaccination_df.merge(countries_flattened, 
                                                left_on='COUNTRY', 
                                                right_on='Country/Region')
vaccination_countries_df
```

```python
# Check the structure of the new dataframe
vaccination_countries_df.info()
```

```python
# Drop the 'Country/Region' column
# vaccination_countries_df = vaccination_countries_df.drop('Country/Region',axis=1)
# vaccination_countries_df
```

```python
# Rename 'Country' to 'Country' for easier reference
vaccination_countries_df = vaccination_countries_df.rename(columns={'COUNTRY':'Country'})
vaccination_countries_df
```

```python
GDP_df.info()
```

```python
# Merge 'vaccination_countries_df' and 'GDP_df'
final_df = vaccination_countries_df.merge(GDP_df, 
                                                left_on='Country', 
                                                right_on='Country Name')
final_df
```

```python
final_df.info()
```

```python
# Drop 'Country Name' and 'ISO Code' columns
final_df = final_df.drop(['Country Name','ISO Code'],axis=1)
final_df
```

```python
# Rename 'PERSONS_BOOSTER_ADD_DOSE_PER100' to 'Booster doses per 100 people' and 'Vaccines - Total doses administered per 100 population'
# to 'Total vaccine doses administered per 100 population'
final_df = final_df.rename(columns={'PERSONS_BOOSTER_ADD_DOSE_PER100':'Booster doses per 100 people',
                                    'Vaccines - Total doses administered per 100 population':'Total vaccine doses administered per 100 population'})
final_df
```

```python
final_df['Country'].unique()
```

```python
# Sort dataframe based on the 'Country' column
final_sorted_df = final_df.sort_values(by=['Country'])
final_sorted_df.head(10)
```

```python
# Re-order columns
final_sorted_df = final_sorted_df[['Country','2015','2016','2017','2018','2019',
                                   'Booster doses per 100 people','Total vaccine doses administered per 100 population',
                                   'Total Confirmed Cases','Newly Confirmed Cases','Total Deaths','New Deaths',
                                   'Total Recovered Cases','Newly Recovered Cases','Lat','Lon']]
final_sorted_df
```

```python
final_sorted_df.info()
```

```python
final_sorted_df['Booster doses per 100 people'] = pd.to_numeric(final_sorted_df['Booster doses per 100 people'], 
                                                                errors='coerce')
final_sorted_df.info()
```

```python
final_sorted_df.to_csv('../Resources/final_sorted_df.csv') 
```
